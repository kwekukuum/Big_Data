**READ DATA**

```python
import pandas as pd


#read .data file into a pandas dataframe

#df = pd.read_csv('adult.data', header = None)
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data', header=None)
df.head()
```

**ASSIGN FEATURE NAMES**

```python
#set variable column_names and assign column names to data
column_names = ['age','workclass','fnlwgt','education','education-num', 'marital-status','occupation','relationship','race','sex','capital-gain','capital-loss', 'hrs-per-wk','native-country','income']

df.columns = column_names

#alternatively
#df.columns=['age','workclass'.....]
```

**PLOT SAMPLE FEATURES WITH HISTOGRAM**

```python
#
import matplotlib.pyplot as plt


sample_features = ['age', 'hrs-per-wk','capital-gain']
df[sample_features].hist()
plt.suptitle('distribution of age, hours per week and capital gain')
plt.show
# title: adds a title to the plot.
# xlabel/ylabel: adds labels to the x and y axes, respectively.
```

**FIND MEAN AND VARIANCE**

```python
#show mean and variance using themean and variance methods
means = df[sample_features].mean().round(2)
variances = df[sample_features].var()

print(means)
print(variances)
```

**STANDARDIZATION**

```python
#extractration of non numerical data or category data types
import numpy as np


category_data = df.select_dtypes(exclude=[np.number]).columns
df[category_data].head()
```

```python
#Label econding
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()

for i in category_data:
    df[i] = label_encoder.fit_transform(df[i])


df.head(20)
```

```python
#Performing Onehot Encoding
from sklearn.preprocessing import OneHotEncoder

onehot_encod = OneHotEncoder(sparse=False)
onehot_data = onehot_encod.fit_transform(df[category_data])
```

```python
#convert to dataframe mapping feature names to data 
onehot_to_dataframe = pd.DataFrame(onehot_data, columns = onehot_encod.get_feature_names_out(category_data))

onehot_to_dataframe
```
